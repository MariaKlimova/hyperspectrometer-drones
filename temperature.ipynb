```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
alum = {
    '1':[
            {
                'h': 55,
                'v': 44,
                'temp': -20
            },
            {
                'h': 60,
                'v': 64,
                'temp': 17
            },
        ],
    '2':[
            {
                'h': 46,
                'v': 64,
                'temp': -20
            },
            {
                'h': 65,
                'v': 64,
                'temp': 17
            },
        ],
        
    '3':[
            {
                'h': 36,
                'v': 41,
                'temp': -20
            },
            {
                'h': 63,
                'v': 56,
                'temp': 17
            },
        ],
    '4':[
            {

                'h': 32,
                'v': 33,
                'temp': -20

            },
            {
                'h': 58,
                'v': 47,
                'temp': 17
            }
        ],
    '5':[
            {
                'h': 54,
                'v': 66,
                'temp': -20
            },
            {
                'h': 73,
                'v': 71,
                'temp': 17
            }
        ]
}
```

```python
fig1, ax1 = plt.subplots(1)
x = [el['temp'] for el in alum['1']]
plt.title('Horizontal')
mira_1_h = [el['h'] for el in alum['1']]
mira_2_h = [el['h'] for el in alum['2']]
mira_3_h = [el['h'] for el in alum['3']]
mira_4_h = [el['h'] for el in alum['4']]
mira_5_h = [el['h'] for el in alum['5']]
ax1.plot(x, mira_1_h, color='b', label='1')
ax1.plot(x, mira_2_h, color='g', label='2')
ax1.plot(x, mira_3_h, color='r', label='3')
ax1.plot(x, mira_4_h, color='m', label='4')
ax1.plot(x, mira_5_h, color='c', label='5')
ax1.legend()
```

```python
fig1, ax1 = plt.subplots(1)
x = [el['temp'] for el in alum['1']]
mira_1_v = [el['v'] for el in alum['1']]
plt.title('Vertical')
mira_1_v = [el['v'] for el in alum['1']]
mira_2_v = [el['v'] for el in alum['2']]
mira_3_v = [el['v'] for el in alum['3']]
mira_4_v = [el['v'] for el in alum['4']]
mira_5_v = [el['v'] for el in alum['5']]
ax1.plot(x, mira_1_v, color='b', label='1')
ax1.plot(x, mira_2_v, color='g', label='2')
ax1.plot(x, mira_3_v, color='r', label='3')
ax1.plot(x, mira_4_v, color='m', label='4')
ax1.plot(x, mira_5_v, color='c', label='5')
ax1.legend()
```
