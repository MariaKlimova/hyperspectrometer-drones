```python
temps = [-20,0,20,30,50]
old = []
new = []
for temp in temps:
    sum_old = 0
    sum_new = 0
    for i in range(1,6):
        for el in alum[str(i)]:
            
            if (el['temp']==temp):
                sum_old += el['h']+el['v']
        for el in alum2[str(i)]:
            
            if (el['temp']==temp):
                sum_new += el['h']+el['v']
    old.append(sum_old//5)
    new.append(sum_new//5)
old,new
```

```python
fig1, ax1 = plt.subplots(1)
x = temps
plt.title('Total')
fig1.set_size_inches(15,10)
ax1.plot(x, old, color='b', label='old')
ax1.plot(x, new, color='r', label='new')
ax1.legend()
```
