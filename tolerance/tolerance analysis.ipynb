```python
x3_arrays = []
y_arrays = []
for e in x:
    num_of_el = x.index(e)
    already_there = False
    for ar in x3_arrays:
        num_of_ar = x3_arrays.index(ar)
        if (e[0] == ar[0][0] and e[1] == ar[0][1]):
            ar.append(e)
            y_arrays[num_of_ar].append(y[num_of_el])
            already_there = True
    if not already_there:
        new_ar = []
        new_ar.append(e)
        x3_arrays.append(new_ar)
        
        new_y_ar = []
        new_y_ar.append(y[num_of_el])
        y_arrays.append(new_y_ar)
        
```
